**Outlier Detection Analysis**

```python
# numerical features
num_cols_outlier= data.select_dtypes(include= ['int64', 'float64'])
```

```python
# Outlier analysis using boxplot
fig, ax = plt.subplots(6, 3, figsize = (16, 14))
ax = ax.flatten()

for i, c in enumerate(num_cols_outlier):
    sns.boxplot(x = num_cols_outlier[c], ax = ax[i], palette = 'Set2')
plt.suptitle('Outlier analysis using Boxplots', fontsize = 12)
plt.tight_layout()
plt.show()
```

```python
# Percentage of outliers present
Q1 = num_cols_outlier.quantile(0.25)
Q3 = num_cols_outlier.quantile(0.75)
IQR = Q3 - Q1
pd.DataFrame((((num_cols_outlier < (Q1 - 1.5 * IQR)) | (num_cols_outlier > (Q3 + 1.5 * IQR))).sum()/num_cols_outlier.shape[0]*100).round(2),
                    columns = ['Outliers %']).sort_values(by = 'Outliers %', ascending = False)
```

```python
# Create another dataset for outlier treated features
data_outlier_treated= data.copy()
```

```python
# Function for outlier treatment

def remove_outlier(col):
  sorted(col)
  Q1, Q3= np.percentile(col, [25, 75])
  IQR = Q3 - Q1
  lower_range= Q1 - 1.5 * IQR
  upper_range= Q3 + 1.5 * IQR
  return lower_range, upper_range
```

```python
# Treating outliers
for col in num_cols_outlier:
    lr, ur = remove_outlier(data_outlier_treated[col])
    data_outlier_treated[col] = np.where(data_outlier_treated[col] > ur, ur, data_outlier_treated[col])
    data_outlier_treated[col] = np.where(data_outlier_treated[col] < lr, lr, data_outlier_treated[col])
```



*   Boxplot of the numerical features showed the presence of outlier in most of the numerical features
*   Out of all features Lot Area (in Sqft) and Lot Area after Renovation (in Sqft) had maximum outliers of approximately 11% and 10% respectively
*   Model is well trained if we look into the dataset with and without outliers.Determining whether to treat outliers is contextual decision based on their nature and business understanding.
*   Hence a new dataset created with outlier treated features
