##### 5. Perform meaningful visualizations using the dataset. Bring at least 3 meaningful visualizations.

```python
# Histogram
fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(10, 10))

plt.suptitle('Histogram of numerical columns')
sns.histplot(df_iris.SL, ax = ax[0,0], kde = True, color = 'lightcoral')
ax[0, 0].set_title('Sepal Length')
sns.histplot(df_iris.SW, ax = ax[0,1], kde = True, color = 'lightsalmon')
ax[0, 1].set_title('Sepal Width')
sns.histplot(df_iris.PL, ax = ax[1,0], kde = True, color = 'lightgreen')
ax[1, 0].set_title('Petal Length')
sns.histplot(df_iris.PW, ax = ax[1,1], kde = True)
ax[1, 1].set_title('Petal Width')

plt.show()
```

Analyzing the histogram we can see the column SW follows a normal distribution whereas the other 3 columns have an asymmetrical distribution.

```python
# Boxplot
fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(10, 10))

plt.suptitle('Boxplot of numerical columns')
sns.boxplot(df_iris.SL, ax = ax[0,0], color = 'lightcoral')
ax[0, 0].set_title('Sepal Length')
sns.boxplot(df_iris.SW, ax = ax[0,1], color = 'lightsalmon')
ax[0, 1].set_title('Sepal Width')
sns.boxplot(df_iris.PL, ax = ax[1,0], color = 'lightgreen')
ax[1, 0].set_title('Petal Length')
sns.boxplot(df_iris.PW, ax = ax[1,1])
ax[1, 1].set_title('Petal Width')

plt.show()
```

A study of boxplots shows presence of outliers in the attribute SW

```python
# Scatter plot
plt.figure(figsize = (6, 5))

sns.scatterplot(x = df_iris.SL, y = df_iris.SW)
plt.title('Scatterplot of SW vs SL')
plt.xlabel('Sepal length (SL)')
plt.ylabel('Sepal width (SW)')
plt.show()
```

A scatter plot between SW and SL shows data points are scattered randomly with no noticeable pattern. It suggests there is no significant relationship between the variables.

```python
plt.figure(figsize = (6, 5))

sns.scatterplot(x = df_iris.PL, y = df_iris.PW)
plt.title('Scatterplot of PW vs PL')
plt.xlabel('Petal length (PL)')
plt.ylabel('Petal width (PW)')
plt.show()
```

Scatterplot between PW and PL show a strong positive relationship between the variables. As PL increases, PW also increases and vice versa.
